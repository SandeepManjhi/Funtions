Q.1. What is the difference between a function and a method in Python?
  - A function is a standalone block of code that can be called multiple times from different parts of a program. A method, on the other hand, is a function that is defined inside a class and is used to perform operations on an object of that class.

Example:
    # Function
    def greet(name):
        print("Hello, {name}!")

    # Method
    class Person:
        def _init_(self, name):
            self.name = name

        def greet(self):
            print("Hello, my name is {self.name}!")

    person = Person("John")
    person.greet()  
    # Output: Hello, my name is John!


Q.2. Explain the concept of function arguments and parameters in Python.
  - Function arguments are the values passed to a function when it is called, while function parameters are the variables defined in the function signature that receive these values.

Example:
    
    def greet(name):  # name is a parameter
        print(f"Hello, {name}!")

    greet("John")  # "John" is an argument


Q.3. What are the different ways to define and call a function in Python?
  - There are several ways to define and call a function in Python, including:

- Defining a function using the def keyword
- Calling a function using its name followed by parentheses
- Using lambda functions
- Using function pointers

Example:
    # Defining a function
    def add(x, y):
        return x + y

    # Calling a function
    result = add(2, 3)
    print(result)  # Output: 5

    # Lambda function
    add_lambda = lambda x, y: x + y
    result = add_lambda(2, 3)
    print(result)  # Output: 5

Q.5. What are iterators in Python and how do they differ from iterables?
  - An iterable is an object that can be iterated over, such as a list or a string. An iterator is an object that keeps track of its position in an iterable and returns the next value each time it is called.

Example:
    
    my_list = [1, 2, 3]
    my_iter = iter(my_list)

    print(next(my_iter))  # Output: 1
    print(next(my_iter))  # Output: 2
    print(next(my_iter))  # Output: 3


Q.6. Explain the concept of generators in Python and how they are defined.
  - A generator is a special type of iterator that can be used to generate a sequence of values on the fly.

Example:

    def my_gen():
        
        yield 1
        yield 2
        yield 3

    my_iter = my_gen()

    print(next(my_iter))  # Output: 1
    print(next(my_iter))  # Output: 2
    print(next(my_iter))  # Output: 3


Q.7. What are the advantages of using generators over regular functions?
  - Generators have several advantages over regular functions, including:

- Memory efficiency: Generators use less memory than regular functions because they only store the current value.
- Flexibility: Generators can be used to implement complex logic and can be composed together.

Example:
    
    def infinite_sequence():
        num = 0
        while True:
            yield num
            num += 1

    seq = infinite_sequence()
    for _ in range(10):
        print(next(seq))  
        # Output: 0, 1, 2, ..., 9

Q.9. Explain the purpose and usage of the map() function in Python.
  - The map() function applies a given function to each item of an iterable and returns a map object.

Example:
    
    def square(x):
        return x ** 2

    numbers = [1, 2, 3, 4, 5]
    squares = list(map(square, numbers))
    print(squares)  
    # Output: [1, 4, 9, 16, 25]


Q.10. What is the difference between 'map()", "reduce()", and filter() functions in Python?
  - map() applies a function to each item of an iterable and returns a map object.
  - reduce() applies a function to the first two items of an iterable, then to the result and the next item, and so on.
  - filter() constructs an iterator from elements of an iterable for which a function returns true.

Example:
    
    def square(x):
        return x ** 2

    numbers = [1, 2, 3, 4, 5]
    squares = list(map(square, numbers))
    print(squares)

```python
# Q.1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
    sum_even = 0
    for num in numbers:
        if num % 2 == 0:
            sum_even += num
    return sum_even
a = input("Enter a list of numbers separated by spaces: ").split()
a = [int(num) for num in a]
sum_even_numbers(a)

```

```python
# Q.3 Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    squares = []
    for num in numbers:
        squares.append(num ** 2)
    return squares
a = input("Enter a list of numbers separated by spaces: ").split()
a = [int(num) for num in a]
square_numbers(a)
```

```python
# Q.4 Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    if num <= 1:
        return False
    if num <= 3:
        return True
    if num % 2 == 0 or num % 3 == 0:
        return False
    i = 5
    while i * i <= num:
        if num % i == 0 or num % (i + 2) == 0:
            return False
        i += 6
    return True

def check_primes_in_range(limit):
    for number in range(1, limit + 1):
        if is_prime(number):
            print(f"{number} is a prime number.")
        else:
            print(f"{number} is not a prime number.")
check_primes_in_range(200)
```

```python
# Q.5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci(n):

    l = [0,1]
    for i in range(2,n):
        l.append(l[-1]+l[-2])
    return l

if __name__ == "__main__":

    n = int(input("Enter the number of terms: "))

    fibo = fibonacci(n)

    print("Fibonacci Series: ",*fibo)
```

```python
# Q.6 Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(n):
    for i in range(n):
        yield 2 ** i

if __name__ == "__main__":
    n = int(input("Enter the exponent: "))

    powers = power_of_two(n)

    for power in powers:
        print(power)
```

```python
# Q.7 Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines(filename):
  with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
a = read_lines("file.txt")
for line in a:
    print(line)
```

```python
# Q.8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
a = [(1, 4), (4, 1), (2, 2), (5,5)]

b = sorted(a, key=lambda x: x[1])

print(b)
```

```python
# Q.10 Create a Python program that uses filter() to remove all the vowels from a given string.
def is_vowel(char):
    return char in "aeiouAEIOU"
a = input("Enter a string: ")
c = ''.join(filter(is_vowel, a))
print(c)
```
